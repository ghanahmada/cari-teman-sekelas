```python
import pandas as pd

import warnings
warnings.filterwarnings("ignore")
pd.set_option("display.max_columns", None)
```

```python
data_23_ilkom = pd.read_excel("data/Pembagian PA Maba 2023.xlsx", "IK")
data_23_ilkom.insert(3, "Prodi", ["IK" for _ in range(len(data_23_ilkom))])

data_23_sisfor = pd.read_excel("data/Pembagian PA Maba 2023.xlsx", "SI")
data_23_sisfor.insert(3, "Prodi", ["SI" for _ in range(len(data_23_sisfor))])

data_23 = pd.concat([data_23_ilkom, data_23_sisfor])
data_23 = data_23.fillna("-").reset_index(drop=True)
data_23.rename(columns={"MD 1": "Matdis 1", "Nama": "Nama Mahasiswa"}, inplace=True)

data_23
```

```python
data_23.info()
```

```python
data_23["Nama Mahasiswa"] = data_23["Nama Mahasiswa"].apply(lambda name: " ".join([w.capitalize() for w in name.split()]))
data_23
```

```python
data_23.to_csv("data/processed_data_pembagian_kelas_2023.csv", index=False)
```
